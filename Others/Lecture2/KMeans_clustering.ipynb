```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
X= 1-3 * np.random.rand(100,2)
X1 = 2 + 4 * np.random.rand(50,2)
X[50:100, :] = X1
plt.scatter(X[ : , 0], X[ :, 1], s = 30, c = 'r')
plt.show()
```

```python
KMean = KMeans(n_clusters=2,n_init=10)
KMean.fit(X)
```

```python
KMean.cluster_centers_
```

```python
plt.scatter(X[ : , 0], X[ : , 1], s =50, c='b')
plt.scatter(4.13310498, 4.0269406, s=60, c='g', marker='s')
plt.scatter(-0.48435662, -0.47876544, s=60, c='r', marker='s')
plt.show()
```

```python
KMean.labels_
```

```python
sample_test=np.array([-3.0,-3.0])
second_test=sample_test.reshape(1, -1)
KMean.predict(second_test)
```
