```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
iris = load_iris()
```

```python
Data = pd.DataFrame(iris.data, columns=iris.feature_names)
Data.head() #Displays first 5 samples
```

```python
Data.describe() #Displays data for every feature
```

```python
plt.scatter(Data['sepal length (cm)'],Data['sepal width (cm)'], s=60)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
```

```python
#Display 3rd and 4th feature
plt.scatter(Data['petal length (cm)'],Data['petal width (cm)'],s=60)
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
```

```python
x = Data.values
kmeans = KMeans(n_clusters=3, max_iter=100,random_state=0,n_init =100)
y_kmeans = kmeans.fit_predict(x)
```

```python
kmeans.cluster_centers_
```

```python
#plots the clusters of data (FOR FIRST 2 FEATURES)
plt.scatter(x[y_kmeans == 0,0],x[y_kmeans == 0,1],s=80,c='red',label='cluster1')
plt.scatter(x[y_kmeans == 1,0],x[y_kmeans == 1,1],s=80,c='blue',label='cluster2')
plt.scatter(x[y_kmeans == 2,0],x[y_kmeans == 2,1],s=80,c='green',label='cluster3')

#plots the centroids
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=80,c='black',label='Centroids',marker = 'x')

#Shows visualy
plt.legend()
```

```python
#plots the clusters of data (FOR LAST 2 FEATURES)
plt.scatter(x[y_kmeans == 0,2],x[y_kmeans == 0,3],s=80,c='red',label='cluster1')
plt.scatter(x[y_kmeans == 1,2],x[y_kmeans == 1,3],s=80,c='blue',label='cluster2')
plt.scatter(x[y_kmeans == 2,2],x[y_kmeans == 2,3],s=80,c='green',label='cluster3')

#plots the centroids
plt.scatter(kmeans.cluster_centers_[:,2],kmeans.cluster_centers_[:,3],s=80,c='black',label='Centroids',marker = 'x')

#Shows visualy
plt.legend()
```
